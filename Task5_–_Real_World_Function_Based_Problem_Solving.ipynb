```python
#Problem 1: Social Media – Post Engagement Analyzer
def analyze_post_engagement(likes_list):
    total_likes = 0

    for like in likes_list:
        total_likes += like

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", total_likes)
    print("Post Status:", status)


# New Example
likes = [120, 340, 560, 150]
analyze_post_engagement(likes)
```

```python
#Problem 2: Healthcare – Medicine Stock Alert System
def check_medicine_stock(stock):

    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"

    print("Medicine Stock:", stock)
    print("Status:", status)


# New Example
check_medicine_stock(15)
```

```python
#Problem 3: Agriculture – Rainfall Adequacy Checker
def rainfall_checker(rainfall_list, required_level):

    total = 0

    for rain in rainfall_list:
        total += rain

    average = total / len(rainfall_list)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", average)
    print("Rainfall Status:", status)


# New Example
rainfall_data = [50, 65, 70, 80, 60]
rainfall_checker(rainfall_data, 65)
```

```python
#Problem 4: Social Media – Duplicate Account Detection
def detect_duplicate_accounts(usernames):

    unique_users = set(usernames)

    if len(unique_users) != len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# New Example
users = ["kiran99", "sneha21", "rohit77", "sneha21"]
detect_duplicate_accounts(users)
```

```python
#Problem 5: Healthcare – Appointment Eligibility Checker
def check_appointment_eligibility(age):

    if age >= 18:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Patient Age:", age)
    print("Eligibility Status:", status)


# New Example
check_appointment_eligibility(16)
```

```python
#Problem 6: Agriculture – Premium Crop Price Filter
def filter_premium_crops(price_list):

    premium = []

    for price in price_list:
        if price > 2000:
            premium.append(price)

    print("Premium Crops:", premium)


# New Example
crop_prices = [2200, 1800, 3400, 1500, 2750]
filter_premium_crops(crop_prices)
```

```python
#Problem 7: System Monitoring – Application Health Checker
def check_system_health(error_count):

    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print("Error Count:", error_count)
    print("System Status:", status)


# New Example
check_system_health(3)
```

```python
#Problem 8: Banking – Daily Transaction Limit Checker
def transaction_checker(amount):

    daily_limit = 50000

    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"

    print("Transaction Amount:", amount)
    print("Transaction Status:", status)


# New Example
transaction_checker(42000)
```

```python
#Problem 9: E-Learning – Student Attendance Eligibility System
def attendance_checker(attendance_list):

    total_classes = len(attendance_list)
    attended = 0

    for status in attendance_list:
        if status == "Present":
            attended += 1

    percentage = (attended / total_classes) * 100

    if percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", eligibility)


# New Example
attendance_data = ["Present", "Present", "Absent", "Present", "Absent", "Present"]
attendance_checker(attendance_data)
```

```python
#Problem 10: Smart Electricity Bill Analyzer
def electricity_bill_calculator(units):

    if units <= 100:
        bill = units * 3

    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)

    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)


    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    print("Units Consumed:", units)
    print("Total Bill:", bill)
    print("Usage Status:", usage)


# New Example
electricity_bill_calculator(180)
```
