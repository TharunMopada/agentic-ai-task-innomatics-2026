Q1. Difference between "Love" and "love" in NLP
"Love" and "love" are treated as different tokens if case normalization is not
applied.
This can increase vocabulary size unnecessarily.
That’s why we usually convert everything to lowercase.

**Example**: If we don’t change everything to lowercase, the computer thinks "Love" and "love" are two totally different topics, which splits its focus and makes it less smart.



Q2. What happens if stopwords are not removed?
Model gets extra noise (words like “is”, “the”, “and”).
Slower processing and less accuracy.
Important words may get less importance.

Example: If you don't remove them, the computer spends all its energy looking at these common words instead of the important ones (like pizza, delicious, slow, broken).




Q3. When removing stopwords is harmful

Sometimes, those "tiny" words carry the whole meaning of the sentence.
Scenario A: Sentiment (Feelings).
Sentence: "I am not happy."
If you remove the stopword "not," the computer thinks you said: "I am happy." It gets the exact opposite meaning!

Scenario B: Searching/Asking Questions.
Sentence: "Who is the president?"
If you remove "Who," the computer just sees "president." It doesn’t know if you’re asking who they are, where they live, or how to become one.



Q4. Stemming vs Lemmatization

Stemming:

Cuts words roughly.

Faster.

Example: "running" - "run" (sometimes "runn")


Lemmatization:

Converts to dictionary form more accurate.

Example: "running" -› "run"

````python
#Task 2: Preprocessing Function
import re

def preprocess_text(text):
    if not text or not isinstance(text, str):
        return [], ""

    text = text.lower()

    text = re.sub(r'http\S+|www\S+|https\S+|\S+@\S+', '', text)

    text = re.sub(r'\d+', '', text)

    text = re.sub(r'[^a-z\s]', '', text)

    text = re.sub(r'(.)\1+', r'\1', text)

    text = re.sub(r'\s+', ' ', text).strip()

    tokens = text.split()

    tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]

    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence
````

````python
#Task 3: Stress Testing
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

for sentence in test_sentences:
    tokens, cleaned = preprocess_text(sentence)
    print("\nOriginal:", sentence)
    print("Tokens:", tokens)
    print("Cleaned:", cleaned)

    results.append(tokens)
````

````python
#Task 4: Token Analytics
import numpy as np

for i, tokens in enumerate(results):
    if len(tokens) == 0:
        continue

    total = len(tokens)
    unique = len(set(tokens))
    avg_len = np.mean([len(word) for word in tokens])

    print(f"\nSentence {i+1}")
    print("Total tokens:", total)
    print("Unique tokens:", unique)
    print("Avg token length:", round(avg_len, 2))
````

````python
#Task 5: Frequency Analysis
from collections import Counter

all_tokens = [token for sublist in results for token in sublist]

counter = Counter(all_tokens)

print("Top 10 frequent:", counter.most_common(10))
print("Least 5 frequent:", counter.most_common()[-5:])
````

````python
#Task 6: Full Pipeline
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, cleaned = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(cleaned)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
````
