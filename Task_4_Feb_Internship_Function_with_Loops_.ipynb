```python
#Problem 1: Mobile Recharge Validation System
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        print("Recharge amount must be at least ₹50")
        return False
    elif amount not in valid_plans:
        print("Invalid plan selected.")
        return False
    else:
        print("Recharge Successful!")
        return True


while True:
    recharge_amount = int(input("Enter recharge amount: "))

    if validate_recharge(recharge_amount):
        break
    else:
        print("Please try again.\n")
```

```python
#Problem 2: Inventory Reorder Alert System
def check_inventory(products):
    for item in products:
        if products[item] < 15:
            print(item, ": Reorder Alert")
        else:
            print(item, ": Stock OK")


# Sample Data
inventory = {
    "Rice": 20,
    "Oil": 10,
    "Sugar": 25,
    "Soap": 8
}

check_inventory(inventory)
```

```python
#Problem 3: Student Result Processing System
def check_result(marks):
    total = 0

    for mark in marks:
        total = total + mark

    average = total / len(marks)
    print("Average Marks:", average)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"


# Sample Data
student_marks = [60, 45, 70, 50]
result = check_result(student_marks)
print("Result:", result)
```

```python
#Problem 4: Cab Fare Estimator with Retry Option
def calculate_fare(distance, peak):
    base_fare = 50
    fare = base_fare + (12 * distance)

    if peak == "yes":
        fare = fare + (fare * 0.25)

    return fare


# Retry option
while True:
    km = float(input("Enter distance in km: "))
    peak_hour = input("Is it peak hour? (yes/no): ").lower()

    total_fare = calculate_fare(km, peak_hour)
    print("Total Fare: ₹", total_fare)

    choice = input("Do you want to calculate again? (yes/no): ").lower()
    if choice != "yes":
        break
```

```python
#Problem 6: Password Strength Checker
def check_password(password):
    has_digit = False
    has_special = False

    special_characters = "@#$"

    if len(password) < 8:
        return "Password too short"

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_characters:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"


# Sample Input
pwd = input("Enter password: ")
print(check_password(pwd))
```

```python
#Problem 5: Employee Attendance Eligibility Checker
def check_attendance(attendance_list):
    present_days = 0

    for day in attendance_list:
        if day == "P":
            present_days = present_days + 1

    percentage = (present_days / len(attendance_list)) * 100
    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


# Sample Data
attendance = ["A", "P", "A", "A", "P", "P", "P", "P"]
status = check_attendance(attendance)
print("Status:", status)
```
