# Deep Technical Guide to LangChain: Building LLM-Powered Applications


# 1. Introduction to LangChain

**What is LangChain?**

LangChain is a framework that helps developers build applications using Large Language Models (LLMs). Instead of calling an LLM directly again and again, LangChain helps you organize and connect different parts like prompts, memory, tools, and APIs into one structured pipeline.

**Why is it Important?**

When working with LLMs directly:

*   You write prompts manually every
time
*   No memory of previous conversations
*   Hard to connect external tools (like search, calculator, APIs)

**LangChain solves this by:**

*   Creating reusable pipelines
*   Managing conversation memory
*   Allowing integration with external tools

**Problems it Solves**

*    Orchestration of multiple steps

*    Memory handling

*   Tool integration (APIs, search, DB)

*   Modular design for scaling applications

# 2. Core Components of LangChain
**2.1 LLMs and Chat Models
Concept**

*  LLMs are the core engines that generate responses.

Why it exists

*  To process natural language and generate human-like text.



```python
#Basic LLM Call
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model="gpt-3.5-turbo", temperature=0.7)

response = llm.invoke("Explain LangChain in simple terms")
print(response.content)
```

**2.2 Prompt Templates
Concept**

*   Instead of writing prompts manually, templates allow dynamic input.

Why it exists

*  To make prompts reusable and structured.

```python
#Prompt Template
from langchain.prompts import PromptTemplate

template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple words"
)

prompt = template.format(topic="LangChain")
print(prompt)
```

```python
#Simple Chain
from langchain.chains import LLMChain

chain = LLMChain(llm=llm, prompt=template)

result = chain.invoke({"topic": "AI Agents"})
print(result["text"])
```

```python
#Memory Example
from langchain.memory import ConversationBufferMemory
from langchain.chains import ConversationChain

memory = ConversationBufferMemory()

conversation = ConversationChain(
    llm=llm,
    memory=memory
)

print(conversation.run("Hi, my name is Tharun"))
print(conversation.run("What is my name?"))
```

```python
#Agent with Tool
from langchain.agents import initialize_agent, Tool
from langchain.agents import AgentType

def calculator_tool(input):
    return str(eval(input))

tools = [
    Tool(
        name="Calculator",
        func=calculator_tool,
        description="Useful for math calculations"
    )
]

agent = initialize_agent(
    tools,
    llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
)

agent.run("What is 25 * 4?")
```

```python
#Document Loader
from langchain.document_loaders import TextLoader

loader = TextLoader("sample.txt")
documents = loader.load()

print(documents[0].page_content)
```

```python
# Vector Store Example
from langchain.vectorstores import FAISS
from langchain.embeddings import OpenAIEmbeddings

embeddings = OpenAIEmbeddings()

db = FAISS.from_texts(
    ["LangChain is powerful", "It helps build AI apps"],
    embeddings
)

query = "What is LangChain?"
docs = db.similarity_search(query)

print(docs[0].page_content)
```

#Advantages and Limitations
**Advantages**

*  Modular design
*  Easy to build pipelines
*  Supports many integrations
*  Rapid prototyping


**Limitations**

*  Debugging is complex
*  Slower due to multiple steps
*  Cost increases with API calls

**When NOT to Use LangChain**

*  Simple one-time LLM calls
*  Small scripts without workflows
