```python
#Problem 1: Smart Parking Lot Management System
def parking_system(capacity, logs):
    parked = 0

    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT":
            if parked > 0:
                parked -= 1

    print("Currently Parked Vehicles:", parked)

    if parked > capacity:
        print("Parking Status: Full")
    else:
        print("Parking Status: Available")


# Example
capacity = 10
logs = ["IN", "IN", "OUT", "IN", "IN"]

parking_system(capacity, logs)
```

```python
#Problem 2: Online Food Delivery Time Estimator
def delivery_time(distance, traffic, weather):
    time = distance * 6   # 6 minutes per km

    if traffic == "High":
        time += 20
    elif traffic == "Medium":
        time += 10

    if weather == "Rainy":
        time += 15

    print("Estimated Delivery Time:", time, "minutes")


# New Example
delivery_time(5, "Medium", "Rainy")
```

```python
#Problem 3: Movie Theatre Seat Occupancy Analyzer
def theatre_status(total_seats, booked_list):
    booked = len(booked_list)
    percentage = (booked / total_seats) * 100

    print("Occupancy:", int(percentage), "%")

    if percentage == 100:
        print("Show Status: Housefull")
    elif percentage >= 70:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")


# New Example
total_seats = 120
booked_list = [1] * 90

theatre_status(total_seats, booked_list)
```

```python
#Problem 4: Cloud Server Load Classification System
def server_status(cpu_list):
    total = 0

    for value in cpu_list:
        total += value

    average = total / len(cpu_list)

    print("Average CPU Load:", int(average), "%")

    if average < 50:
        print("Server Status: Normal")
    elif average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# New Example
cpu_values = [30, 55, 60, 75]

server_status(cpu_values)
```

```python
#Problem 5: Smart Classroom Resource Usage Monitor
def classroom_monitor(resources):
    overused = []

    for item in resources:
        if resources[item] > 7:
            overused.append(item)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# New Example
usage_data = {
    "Projector": 5,
    "AC": 8,
    "Lights": 9
}

classroom_monitor(usage_data)
```

```python
#Problem 6: Online Event Registration Capacity Controller
def registration_system(capacity, total_registered):

    if total_registered > capacity:
        confirmed = capacity
        waitlist = total_registered - capacity
        status = "Closed"
    else:
        confirmed = total_registered
        waitlist = 0
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)


# New Example
registration_system(50, 65)
```
